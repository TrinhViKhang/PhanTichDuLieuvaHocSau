```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['JavaScript', 'SQL', 'Java', 'C#', 'Python', 'PHP', 'C++', 'C', 'TypeScript', 'Ruby']
students = [62, 51, 39, 34, 32, 28, 22, 19, 9, 9]
ax.bar(langs,students)
plt.show()
```

BĐ Tròn

```python
from matplotlib import pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
langs = ['JavaScript', 'SQL', 'Java', 'C#', 'Python', 'PHP', 'C++', 'C', 'TypeScript', 'Ruby']
students = [62, 51, 39, 34, 32, 28, 22, 19, 9, 9]
ax.pie(students, labels = langs,autopct='%1.2f%%')
plt.show()
```
